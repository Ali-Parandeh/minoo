# SCORE CALCULATION

```python
categories
```

```python
categories.index('x') if "x" in categories else None
```

```python
def score(raw_df: pd.DataFrame, category: str) -> pd.DataFrame:
    if category not in categories:
        return pd.DataFrame()
    raw_df = raw_df[raw_df.columns[raw_df.columns.str.contains(category)]].copy()
    raw_df[f'{category}_avg'] = raw_df.replace(9, np.nan).mean(axis=1)
    raw_df[f'{category}_raw'] = raw_df[f'{category}_avg'] * counts[categories.index(category)]
    raw_df[f'{category}_norm'] = (round(abs(raw_df[f'{category}_raw'] - raw_df[f'{category}_raw'].max()) / ranges[categories.index(category)], 2) * 100).fillna(999).astype(int)
    return raw_df
```

```python
final_df = df[['PIN']].copy()
for cat in categories:
    scale = score(df, cat)
    final_df = pd.concat([final_df, scale], axis=1)
```

# PROCESS ALL DATA

```python
# Define sheet mapping
sheet_mapping = {0: "V2 Placebo", 1: "V2 Liraglutide", 2: "V6 Placebo", 3: "V6 Liraglutide"}
```

```python
# Category Data
categories = ['p', 'sf', 'sx', 'd', 'w']
counts = [11, 7, 4, 5, 4]
min_answered = [6, 4, 2, 3, 2]
ranges = [(5-1) * i for i in counts]

p_cols = ['PIN']
for cat, count in zip(categories, counts):
    p_cols += [cat + str(i) for i in range(0, count)]

strings = ['NEVER TRUE', 'RARELY TRUE', 'SOMETIMES TRUE', 'OFTEN TRUE', 'USUALLY TRUE', 'ALWAYS TRUE', np.nan]
replacements = [1, 2, 3, 4, 4, 5, 9]
```

```python
for k, v in sheet_mapping.items():
    df = pd.read_excel('data.xlsx', sheet_name=k)

    if 'Physical Function  ' in df.columns:
        df = df.drop(columns=['Physical Function  ', 'Self-esteem', 'Sexual Life ', 'Public Distress ', 'Work'])

    df.columns = p_cols

    for s, r in zip(strings, replacements):
        df = df.replace(s, r)

    final_df = df[['PIN']].copy()
    for cat in categories:
        scale = score(df, cat)
        final_df = pd.concat([final_df, scale], axis=1)

    final_df.to_csv(f'{v}.csv', index=False)
```
